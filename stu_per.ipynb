```python
#Convert Categorical data to Numerical data
!pip install scikit-learn
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
df["Extracurricular_Activity"] = le.fit_transform(df["Extracurricular_Activity"])
# it's convert into numerical form
```

```python
df.head()
# NOW entire dataset is in numerical form
```

```python
x = df[["Hours_Studied", "Previous_Score", "Extracurricular_Activity", "Sleep_Hours", "Sample_QP_Practiced"]]
```

```python
x.head()
```

```python

y = df["Performance_Index"]
```

```python
y.head()
```

```python
#create feature(x) , and label(y)  ...  
#split data in train data and test data ...  
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2)
```

```python
x_train
```

```python
x_test
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
# it is convert into Statndard form where u=0 and SD=1
```

```python
from sklearn.linear_model import LinearRegression
import pickle  # to store physical form file for deployment
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
# to predict data
y_pred = model.predict(x_test_scaled)
```

```python
x_test_scaled
```

```python
model.predict([[-1.50840652, -0.5593056 , -0.85096294, -0.22900614, -1.29974735]])
# Here predict the x value and get value of Y
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test,y_pred)
#94% accuracy
```

```python
import pickle
with open("student_lr_final_model.pkl", 'wb') as file:
    pickle.dump((model,scaler,le),file)
# Now model store in physical file...
```
